```python
def hashtag_query(query_list, since_time, until_time):
    print("hashtag_query", query_list)
    print ("since: ",since_time)
    print("until:", until_time)
    word_list =[]
    if query_list[0].startswith('#'):
        word= query_list[1]
        hashtag=query_list[0][1:]
    else :
        word= query_list[0]
        hashtag=query_list[1][1:]
    word_list.append(word)
    print("being passed ", hashtag, word_list)
    
    data2 = scrape(words=word_list, hashtag=hashtag, since=since_time, until=until_time, from_account = None,interval=1, 
      headless=True, display_type="Top", save_images=False, proxy = None, save_dir = 'outputs',
             resume=False, filter_replies=True, proximity=False)
    

    return data2
    
```

```python
def word_query(query_list, since_time, until_time):
    joined_string = ",".join(query_list)
    print ("since: ",since_time)
    print("until:", until_time)
    word_list =[]
    word_list.append(joined_string)
    print("joined string",joined_string)
    print("being passed ", word_list)
    
    data2 = scrape(words=word_list, since=since_time, until=until_time, from_account = None,interval=1, 
      headless=True, display_type="Top", save_images=False, proxy = None, save_dir = 'outputs',
             resume=False, filter_replies=True, proximity=False)
    
    return data2
```

```python
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename[:-4] for filename in filenames if filename.endswith( suffix ) ]
```

```python
def write_df_to_file(df,file_name):
    print(file_name)
    df.to_csv(file_name)
```

```python
since_until_dict_per_issue = dict()
```

```python
def populate_time_dict(issue):
    # Start and End date for search query. example : %Y-%m-%d.
    if issue == 'abortion':
        since_until_dict_per_issue[issue] = ['2021-09-01','2021-09-28']
    elif issue == 'climate_change':
         since_until_dict_per_issue[issue] = ['2021-10-31','2021-11-20']
    elif issue == 'covid':
         since_until_dict_per_issue[issue] = ['2020-03-20','2020-03-28']
    elif issue == 'gun_control':
         since_until_dict_per_issue[issue] = ['2019-09-09','2019-09-19']
    elif issue == 'immigration':
         since_until_dict_per_issue[issue] = ['2021-03-02','2021-03-23']
    elif issue == 'racism':
         since_until_dict_per_issue[issue] = ['2020-05-25','2020-06-09']
    else :
        #this is used to test 
        print ("out of time error!")
        since_until_dict_per_issue[issue] = ['2019-09-01','2019-09-02']

```

```python
### Lets go get the queries we need to make from the directories ###
data_dir = "D:/corporate_social_advocacy/bare_hao_CSA_data/"
issues = [name for name in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, name))]
print("actually", issues)
#issues = ['temp']
for issue in issues:
    populate_time_dict(issue)
    since_time, end_time = since_until_dict_per_issue[issue]
    #print("from", since_time ,"to", end_time)
    new_path = data_dir+issue+"/"
    output_path = "D:/corporate_social_advocacy/round2/"+issue+"/"
    onlyfiles = find_csv_filenames(new_path)
    for query in onlyfiles:
        query_list = query.split(' AND ')
        if query_list[0].startswith('#') or query_list[1].startswith('#'):
            print("hashtag_query triggered", query_list)
            df = hashtag_query(query_list, since_time, end_time)
            write_df_to_file (df, output_path+query+".csv")
        else:
            print("word_query triggered", query_list)
            df = word_query(query_list, since_time, end_time)
            write_df_to_file (df, output_path+query+".csv")
            
```
